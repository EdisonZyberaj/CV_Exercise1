```python
# Import libraries

import numpy as np
import matplotlib.pyplot as plt

from skimage import data
from skimage.color import rgb2gray
from skimage.filters import gaussian, laplace
from skimage.feature import canny
from skimage.transform import resize
```

```python
# Load astronaut image

astro_img = data.astronaut()

# Display image

plt.figure(figsize=(8,6))
plt.imshow(astro_img)
plt.title("Original Astronaut Image")
plt.axis("off")
plt.show()
```

```python
# Convert RGB image to grayscale

gray_astro = rgb2gray(astro_img)

# Show grayscale image

plt.figure(figsize=(8,6))
plt.imshow(gray_astro, cmap='gray')
plt.title("Grayscale Astronaut")
plt.axis("off")
plt.show()
```

```python
# Apply Gaussian smoothing

blurred_img = gaussian(gray_astro, sigma=2)

# Display blurred image

plt.figure(figsize=(8,6))
plt.imshow(blurred_img, cmap='gray')
plt.title("Gaussian Blurred Image")
plt.axis("off")
plt.show()
```

```python
# Apply Laplacian edge enhancement

laplace_img = laplace(blurred_img)

# Display result

plt.figure(figsize=(8,6))
plt.imshow(laplace_img, cmap='gray')
plt.title("Laplacian Filter")
plt.axis("off")
plt.show()
```
