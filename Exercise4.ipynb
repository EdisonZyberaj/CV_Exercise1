```python
# Import libraries

import numpy as np
import matplotlib.pyplot as plt

from skimage import data
from skimage.color import rgb2gray
from skimage.filters import gaussian, laplace
from skimage.feature import canny
from skimage.transform import resize
```

```python
# Load astronaut image

astro_img = data.astronaut()

# Display image

plt.figure(figsize=(8,6))
plt.imshow(astro_img)
plt.title("Original Astronaut Image")
plt.axis("off")
plt.show()
```

```python
# Convert RGB image to grayscale

gray_astro = rgb2gray(astro_img)

# Show grayscale image

plt.figure(figsize=(8,6))
plt.imshow(gray_astro, cmap='gray')
plt.title("Grayscale Astronaut")
plt.axis("off")
plt.show()
```

```python
# Apply Gaussian smoothing

blurred_img = gaussian(gray_astro, sigma=2)

# Display blurred image

plt.figure(figsize=(8,6))
plt.imshow(blurred_img, cmap='gray')
plt.title("Gaussian Blurred Image")
plt.axis("off")
plt.show()
```

```python
# Apply Laplacian edge enhancement

laplace_img = laplace(blurred_img)

# Display result

plt.figure(figsize=(8,6))
plt.imshow(laplace_img, cmap='gray')
plt.title("Laplacian Filter")
plt.axis("off")
plt.show()
```

```python
# Detect edges using Canny

canny_edges = canny(blurred_img)

# Show edges

plt.figure(figsize=(8,6))
plt.imshow(canny_edges, cmap='gray')
plt.title("Canny Edge Detection")
plt.axis("off")
plt.show()
```

```python
# Resize image

small_img = resize(gray_astro, (200, 200))

# Display resized image

plt.figure(figsize=(6,6))
plt.imshow(small_img, cmap='gray')
plt.title("Resized Image")
plt.axis("off")
plt.show()
```

```python
# Compare all processing steps

fig, ax = plt.subplots(2, 3, figsize=(15,10))

ax[0,0].imshow(astro_img)
ax[0,0].set_title("Original")
ax[0,0].axis("off")

ax[0,1].imshow(gray_astro, cmap='gray')
ax[0,1].set_title("Grayscale")
ax[0,1].axis("off")

ax[0,2].imshow(blurred_img, cmap='gray')
ax[0,2].set_title("Blurred")
ax[0,2].axis("off")

ax[1,0].imshow(laplace_img, cmap='gray')
ax[1,0].set_title("Laplacian")
ax[1,0].axis("off")

ax[1,1].imshow(canny_edges, cmap='gray')
ax[1,1].set_title("Canny Edges")
ax[1,1].axis("off")

ax[1,2].imshow(small_img, cmap='gray')
ax[1,2].set_title("Resized")
ax[1,2].axis("off")

plt.tight_layout()
plt.show()
```

```python
# Calculate image statistics

print("Image Statistics")
print("-----------------------")

print("Shape:", gray_astro.shape)
print("Minimum pixel value:", np.min(gray_astro))
print("Maximum pixel value:", np.max(gray_astro))
print("Mean pixel value:", np.mean(gray_astro))
```
